```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://datascience.virginia.edu/people/phil-bourne'
```

```python
page = requests.get(url)
```

```python
Phillip = BeautifulSoup(page.text,'html')
```

```python
print(Phillip.prettify())
```

```python
Phillip.find_all('div')
```

```python
Phillip.find("div",class_ = "field--bio field--body").text
```

```python
nealUrl = "http://datascience.virginia.edu/people/daniel-graham"
page2 = requests.get(nealUrl)
Neal = BeautifulSoup(page2.text, 'html.parser')
```

```python
list_of_expertise = Neal.find_all("div", class_ = "list-text")
expert_in = [i.get_text(strip=True) for i in list_of_expertise]
print(expert_in)
```

```python
#trying to pull from the list of faculty from the UVA data science faculty page
Final_url = "https://datascience.virginia.edu/search?t=people&persons=team&page=1"
page = requests.get(Final_url)
names = BeautifulSoup(page.text, 'html')
```

```python
print(names)
```

```python
x = names.find_all(class_ = "field--name")
#print(f"\n\n {x} \n \n")

list_names = []
for i in x:
    list_names.append(i.get_text(strip=True))
    #print(i.get_text(strip=True))
print(list_names)
```
